## Visualization & Explanatory Analysis

### Pertanyaan 1: Berapa rata-rata suhu  dari tahun ke tahun ? 
### 10122360 - Ghamir Diah Kharismasya0

```python
# Mengatur ukuran dan gaya plot
plt.figure(figsize=(10, 6))
sns.barplot(x="year", y="TEMP",data=data_kota,color='tomato')

# Menambahkan judul dan label sumbu
plt.title('Perubahan Rata Rata Suhu Dari Tahun Ke Tahun')
plt.xlabel('Tahun')
plt.ylabel('Suhu Terendah (°C)')

# Menampilkan grafik
plt.show()
```

### Pertanyaan 2:Bagaimana rata-rata suhu terendah per Bulan ?
### 10122387 - Teguh Darmawansya 

```python
# Mengatur ukuran dan gaya plot
plt.figure(figsize=(10, 6))
sns.barplot(x="month", y="TEMP",data=data_kota, color='red')


# Menambahkan judul dan label sumbu
plt.title('Perubahan Suhu setiap bulan')
plt.xlabel('Bulan')
plt.ylabel('Suhu (°C)')


# Menampilkan grafik
plt.show()
```

### Pertanyaan 3: Bagaimana intensitas hujan tertinggi dan terendah dari tahun ke tahun ?
### 10122370 - M Nata Jayndra Permana

```python
# Mengatur ukuran dan gaya plot
plt.figure(figsize=(10, 6))

# Menggambar grafik batang intensitas hujan tertinggi dari tahun ke tahun
sns.barplot(x="year", y="RAIN", data=data_kota.groupby('year')['RAIN'].max().reset_index(), color='yellowgreen')

# Menambahkan judul dan label sumbu
plt.title('Intensitas Hujan Tertinggi per-Tahun Stasiun Aotizhongxin')
plt.xlabel('Tahun')
plt.ylabel('Intensitas Hujan Tertinggi (mm/hari)')

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```

```python
# Mengatur ukuran dan gaya plot
plt.figure(figsize=(10, 6))

#mengatur data intensitas hujan terendah pertahun
sns.barplot(x="year", y="RAIN",data=data_kota, color='yellowgreen')

# Menambahkan judul dan label sumbu
plt.title('Intensitas Hujan Terendah per-Tahun Stasiun Aotizhongxin')
plt.xlabel('Tahun')
plt.ylabel('Intensitas Hujan (mm/hari)')

# Menampilkan grafik
plt.show()
```

### Pertanyaan 4: Bagaimana rata-rata Intensitas hujan Terendah per Bulan pada tahun 2015 di Stasiun Aotizhongxin ?
### 10122370 - M Nata Jayndra Permana 

```python
#Data tahun 2015
data_2015 = data_kota[data_kota['year'] == 2015]

# Mengatur ukuran dan gaya plot
plt.figure(figsize=(10, 6))

# Menggambar grafik garis
sns.lineplot(x="month", y="RAIN", data=data_2015, marker='o', color='Green')

# Menambahkan judul dan label sumbu
plt.title('Intensitas Hujan terendah per bulan Stasiun Aotizhongxin pada tahun 2015')
plt.xlabel('Bulan')
plt.ylabel('Intensitas Hujan (mm/hari)')

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.show()
```

### Pertanyaan 5: Bagaimana rata-rata Tingkat NO2 per Tahun dan per Bulan ?
### 10122372 - Danendra Muzhaffar

### Pertanyaan 6: Bagaimana rata-rata Perubahan Tingkat CO per Tahun Pada tiap Bulannya ?
### 10122358 - Muhammad Irfan Dava Nugraha

```python
# 2013
data_2013 = data_kota[data_kota['year'] == 2013]
plt.figure(figsize=(10, 6))
sns.lineplot(x="month", y="CO", data=data_2013, marker='o', color='Blue')
plt.title('Tingkat CO Ditahun 2013')
plt.xlabel('Bulan')
plt.ylabel('Tingkat CO')
plt.grid(True)
plt.show()
```

```python
# 2014
data_2014 = data_kota[data_kota['year'] == 2014]
plt.figure(figsize=(10, 6))
sns.lineplot(x="month", y="CO", data=data_2014, marker='o', color='Yellow')
plt.title('Tingkat CO Ditahun 2014')
plt.xlabel('Bulan')
plt.ylabel('Tingkat CO')
plt.grid(True)
plt.show()
```

```python
# 2015
data_2015 = data_kota[data_kota['year'] == 2015]
plt.figure(figsize=(10, 6))
sns.lineplot(x="month", y="CO", data=data_2015, marker='o', color='Green')
plt.title('Tingkat CO Ditahun 2015')
plt.xlabel('Bulan')
plt.ylabel('Tingkat CO')
plt.grid(True)
plt.show()
```

```python
# 2016
data_2016 = data_kota[data_kota['year'] == 2016]
plt.figure(figsize=(10, 6))
sns.lineplot(x="month", y="CO", data=data_2016, marker='o', color='Cyan')
plt.title('Tingkat CO Ditahun 2016')
plt.xlabel('Bulan')
plt.ylabel('Tingkat CO')
plt.grid(True)
plt.show()
```

```python
# 2017
data_2017 = data_kota[data_kota['year'] == 2017]
plt.figure(figsize=(10, 6))
sns.lineplot(x="month", y="CO", data=data_2017, marker='o', color='Cyan')
plt.title('Tingkat CO Ditahun 2017')
plt.xlabel('Bulan')
plt.ylabel('Tingkat CO')
plt.grid(True)
plt.show()
```
